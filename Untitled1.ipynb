```python
import numpy as np
import pandas as pd

# Imports para visualizar la información

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

# Imports relacionados a información del mercado

import yfinance as yf

# Import para ignorar ciertas advertencias

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

%matplotlib inline
```

```python
file_path = r'C:\Users\chris\Downloads\sp500_companies.csv'
data = pd.read_csv(file_path)
print(data.head(10))
```

```python
# Cambié algunos nombres de las columnas para que fueran más claros
data.rename(columns={  # Aquí corregimos 'column' a 'columns'
    "Exchange Symbol": "Exchange", 
    "Shortname": "Short_Name",
    "Longname": "Long_Name",
    "Currentprice": "Current_Price",
    "Marketcap": "Market_Cap",
    "Revenuegrowth": "Revenue_Growth",
    "Fulltimeemployees": "Full_Time_Employees",
    "Longbusinesssummary": "Business_Summary"
}, inplace=True)

# Mostrar columnas para verificar los cambios
for idx, col in enumerate(data.columns):
    print(f"{idx}: {repr(col)}")
```

```python
if "Ebitda" in data.columns:
    data['Ebitda'].fillna(0, inplace=True)
else:
    print("La columna 'Ebitda' no existe.")
```

```python
# Detectar columnas categóricas automáticamente
categorical_cols = data.select_dtypes(include='object').columns.tolist()

# Normalizar las columnas categóricas
for col in categorical_cols:
    if col in data.columns:
        data[col] = data[col].str.lower().str.strip()
    else:
        print(f"Columna {col} no encontrada.")
```

```python
missing_columns = set(['Symbol', 'Short_Name', 'Long_Name', 'Sector', 'Industry',
                       'Current_Price', 'Market_Cap', 'Revenue_Growth', 'Ebitda',
                       'City', 'State', 'Country', 'Full_Time_Employees',
                       'Business_Summary', 'Weight']) - set(data.columns)

if not missing_columns:
    data = data[['Symbol', 'Short_Name', 'Long_Name', 'Sector', 'Industry', 'Current_Price',
                 'Market_Cap', 'Revenue_Growth', 'Ebitda', 'City', 'State', 'Country',
                 'Full_Time_Employees', 'Business_Summary', 'Weight']]
else:
    print(f"Faltan las columnas: {missing_columns}")
```

```python
if {"Market_Cap", "Ebitda"}.issubset(data.columns):
    data['Market_Cap'] = data['Market_Cap'] / 1e9
    data['Ebitda'] = data['Ebitda'] / 1e9
    data.rename(columns={"Market_Cap": "Market_Cap_Billions", "Ebitda": "Ebitda_Billions"}, inplace=True)
else:
    print("Faltan las columnas 'Market_Cap' o 'Ebitda'.")
```

```python
# Mantén solo una vez el método drop_duplicates
data.drop_duplicates(inplace=True)

# Validación de precios negativos
negative_prices = data[data['Current_Price'] < 0]
if not negative_prices.empty:
    print("Precios negativos encontrados:", negative_prices)
else:
    print("No se encontraron precios negativos.")
```

```python
# Reemplazar valores en blanco o espacios en "State"
data["State"] = data["State"].replace(r"^\s*$", np.nan, regex=True)

# Diccionario para mapear ciudades a estados
city_to_state = {
    "cupertino": "CA", "redmond": "WA", "austin": "TX", "dublin": "DUBL", "woking": "WOK",
    "zurich": "ZRH", "eindhoven": "EIN", "schaffhausen": "SH", "cork": "CO",
    "swords": "SW", "herzliya": "HZY", "hamilton": "H", "london": "LDN",
}

# Rellenar valores faltantes en "State"
data["State"] = data["State"].fillna(data["City"].map(city_to_state)).fillna("Unknown")

# Comprobaciones finales
print("Valores únicos en 'State':", data["State"].unique())
print("Valores NaN en 'State':", data["State"].isna().sum())
```

```python
# Formatear 'State' en mayúsculas
data["State"] = data["State"].str.upper()

# Análisis de estados desconocidos
unknown_states = data[data["State"] == "UNKNOWN"]
print(f"Estados desconocidos encontrados: {len(unknown_states)}")

# Visualización de la distribución por estado
state_counts = data["State"].value_counts()
state_counts.plot(kind="bar", figsize=(10, 6), title="Distribución por Estado")
plt.xlabel("Estado")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# 8. Verificar datos faltantes en "State"

# Ver cuántos valores faltantes hay en la columna "State"
print(data["State"].isna().sum())

# Filtrar filas con valores NaN en "State"
missing_state_rows = data[data["State"].isna()]
print(missing_state_rows)
```

```python
# Guardar los datos limpios como un nuevo archivo CSV
data.to_csv(r'C:\Users\chris\Proyecto 3\sp500_companies_cleaned.csv', index=False)
print(data.head())
```

```python
# Cargar datos externos
historical_prices = pd.read_csv(r'C:\Users\chris\Proyecto 3\historical_prices.csv')  

#forzar la conversión
historical_prices['Close'] = pd.to_numeric(historical_prices['Close'], errors='coerce')

# Eliminar filas con NaN en la columna 'Close'
historical_prices.dropna(subset=['Close'], inplace=True)

# Calcular 'Price_High' y 'Price_Low' basados en la columna 'Close'
historical_prices['Price_High'] = historical_prices['Close'].max()
historical_prices['Price_Low'] = historical_prices['Close'].min()

# Renombrar 'ticker' a 'Symbol' si es necesario
if 'ticker' in historical_prices.columns:
    historical_prices.rename(columns={'ticker': 'Symbol'}, inplace=True)

# Verifica las columnas de ambos dataframes
print("Columnas de historical_prices:", historical_prices.columns)
print("Columnas de data:", data.columns)

# Extraer el símbolo de la acción del MultiIndex de `data`
symbol = data.columns.get_level_values(1).unique()[0]  # Obtiene el símbolo de la acción

# Resetear el índice de `data` y agregar la columna 'Symbol'
data_reset = data.reset_index()
data_reset['Symbol'] = symbol

# Poner en ambos dataframes tengan la columna 'Symbol' para la fusión
if 'Symbol' in historical_prices.columns and 'Symbol' in data_reset.columns:
    # Fusión de los dataframes
    data_merged = pd.merge(data_reset, historical_prices[['Symbol', 'Price_High', 'Price_Low']], on='Symbol', how='left')

    # Continuar con el cálculo si la fusión fue exitosa
    if 'Price_High' in data_merged.columns and 'Price_Low' in data_merged.columns:
        price_comparison = data_merged[['Symbol', 'Price_High', 'Price_Low']].copy()
        price_comparison['Price_Change'] = price_comparison['Price_High'] - price_comparison['Price_Low']
        price_comparison_sorted = price_comparison.sort_values(by='Price_Change', ascending=False).head(10)
        print(price_comparison_sorted)
    else:
        print("No se encontraron 'Price_High' o 'Price_Low' después de la fusión.")
else:
    print("No se encontró la columna 'Symbol' en ambos dataframes para realizar la fusión.")
```

```python
# Seleccionar los datos para 'AAPL' de `data`
high_prices_aapl = data[('High', 'AAPL')]
low_prices_aapl = data[('Low', 'AAPL')]

# Crear el DataFrame con los precios de 'AAPL'
df_prices_aapl = pd.DataFrame({
    'Symbol': ['AAPL'] * len(high_prices_aapl),  # Asumiendo que todos los datos son de AAPL
    'High Price': high_prices_aapl,
    'Low Price': low_prices_aapl
})

# Calcular el rango de precios (diferencia entre alto y bajo)
df_prices_aapl['Price Range'] = df_prices_aapl['High Price'] - df_prices_aapl['Low Price']

# Mostrar los resultados
print(df_prices_aapl.head())

# Explicación:
# Symbol: La columna "Symbol" tiene el valor AAPL para todas las filas, ya que todos los datos corresponden a las acciones de Apple.

# High Price: La columna "High Price" muestra el precio más alto registrado para las acciones de AAPL en cada fecha.

# Low Price: La columna "Low Price" muestra el precio más bajo registrado para las acciones de AAPL en cada fecha.

# Price Range: La columna "Price Range" es el cálculo de la diferencia entre el precio más alto y el precio más bajo para cada fecha. Esta columna es útil para analizar la volatilidad de las acciones de AAPL en cada día.

# Date (Índice): El índice del DataFrame es la fecha en que se registraron esos precios, con la zona horaria en formato UTC (+00:00), lo cual es típico en series temporales financieras.

# Por ejemplo, para el 2 de enero de 2020:

# High Price fue $75.150002
# Low Price fue $73.797501
# El Price Range es la diferencia entre estos dos valores: $75.150002 - $73.797501 = $1.352501.
```

```python
import matplotlib.pyplot as plt

# Graficar precios altos y bajos
plt.figure(figsize=(10,6))
plt.plot(df_prices_aapl.index, df_prices_aapl['High Price'], label='High Price', color='green')
plt.plot(df_prices_aapl.index, df_prices_aapl['Low Price'], label='Low Price', color='red')
plt.title('High and Low Prices of AAPL')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
